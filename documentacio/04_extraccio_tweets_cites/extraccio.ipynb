```python
import pandas as pd

# Load the CSV file into a pandas DataFrame
df_date = pd.read_csv('data_by_date_html.csv', low_memory=False)
df_id = pd.read_csv('data_by_id_html.csv', low_memory=False)
```

```python
# Let's check the columns of the DataFrames
print(df_date.columns)
print(df_id.columns)
```

```python
# Define the list of desired columns
keep_cols = ['id', 'data', 'entradeta', 'permatitle', 'avantitol', 'tematiques',
             'tags', 'titol', 'titol_portada', 'comarques', 'cos', 'redactor', 'imatges']

# Keep only the columns that exist in both the dataframe and keep_cols
df_date = df_date[[col for col in keep_cols if col in df_date.columns]]
df_id = df_id[[col for col in keep_cols if col in df_id.columns]]
```

```python
# Let's check the columns of the DataFrames
print(df_date.columns)
print(df_id.columns)
```

```python
# Get the set of overlapping columns, excluding 'id'
common_cols = set(df_date.columns).intersection(df_id.columns) - {'id'}

# Drop those overlapping columns (except 'id') from df_date
df_date_unique = df_date.drop(columns=common_cols)

# Merge the two DataFrames on 'id'
merged_df = pd.merge(df_date_unique, df_id, on='id', how='inner')
```

```python
from bs4 import BeautifulSoup, Comment
import pandas as pd

def extract_and_clean_content(html):
    if not isinstance(html, str):
        return "", "", ""

    soup = BeautifulSoup(html, 'html.parser')

    # Extract tweets
    tweets = []
    for tweet in soup.find_all('blockquote', class_='twitter-tweet'):
        tweets.append(tweet.get_text(" ", strip=True))
        tweet.decompose()

    # Extract other quotes
    quotes = []
    for quote in soup.find_all('blockquote'):
        quotes.append(quote.get_text(" ", strip=True))
        quote.decompose()

    # Remove <script> tags
    for script in soup.find_all('script'):
        script.decompose()

    # Remove HTML comments
    for comment in soup.find_all(string=lambda text: isinstance(text, Comment)):
        comment.extract()

    # Remove empty <p>, <div>, etc.
    for tag in soup.find_all(['p', 'div']):
        if not tag.get_text(strip=True):
            tag.decompose()

    # Keep only these tags
    allowed_tags = {'p', 'strong', 'em', 'h1', 'h2', 'h3', 'h4', 'h5', 'h6',
                    'ul', 'ol', 'li', 'a', 'br', 'span'}
    for tag in soup.find_all():
        if tag.name not in allowed_tags:
            tag.unwrap()  # remove tag but keep its content

    return str(soup).strip(), '\n\n'.join(tweets), '\n\n'.join(quotes)

def process_cos_column(df):
    df = df.copy()

    cleaned_cos = []
    tweet_list = []
    quote_list = []

    for html in df['cos']:
        cleaned_html, tweets, quotes = extract_and_clean_content(html)
        cleaned_cos.append(cleaned_html)
        tweet_list.append(tweets)
        quote_list.append(quotes)

    df['cos'] = cleaned_cos
    df['tweets'] = tweet_list
    df['quotes'] = quote_list

    return df

# Apply it to your DataFrame
merged_df = process_cos_column(merged_df)
```

```python
def has_bad_unicode(text):
    if not isinstance(text, str):
        return False
    try:
        text.encode('utf-8')
        return False
    except UnicodeEncodeError:
        return True

# Check how many rows are affected
bad_rows = merged_df.applymap(has_bad_unicode).any(axis=1)
print(f"Rows with bad unicode: {bad_rows.sum()}")

# Optionally, view them
print(merged_df[bad_rows])
```

```python
def clean_surrogates(text):
    if not isinstance(text, str):
        return text
    return text.encode('utf-8', 'replace').decode('utf-8')

# Apply this to the whole DataFrame
merged_df = merged_df.applymap(clean_surrogates)
```
